###### Remove all rows with NULL values:

```python
new_df = df.dropna()

print(new_df.to_string())
```

Note: By default, the dropna() method returns a new DataFrame, and will not change the original.

If we  want to change the original DataFrame, we should use the **inplace = True** argument

###### Replace NULL values with the number 270

```python
new1_df = df.fillna(270) #The fillna() method allows us to replace empty cells with a value
```

```python
print(new1_df)
```

###### Replace NULL values in the "Calories" columns with the number 130:

```python
new_q=df["Calories"].fillna(130)
#To only replace empty values for one column, we can specify the column name for the DataFrame
print(new_q.to_string())
```

###### Calculate the MEAN, and replace any empty values with it:

###### Mean = the average value (the sum of all values divided by number of values).

```python
x=df['Calories'].mean()
```

```python
new_w=df['Calories'].fillna(x)
print(new_w.to_string())
```

###### Calculate the MEDIAN, and replace any empty values with it:
    Median = the value in the middle, after you have sorted all values ascending.

```python
y = df['Calories'].median()
new_e = df['Calories'].fillna(y)
print(new_e.to_string())
```

###### Mode = the value that appears most frequently.
Calculate the MODE, and replace any empty values with it:

```python
z = df['Calories'].mode()[0]
new_r = df['Calories'].fillna(z)
print(new_r)
```

```python
print(df)
```

```python
df.dropna(inplace = True)

print(df.to_string())
```

```python
df.loc[0,'Pulse'] = 130
#Pandas use the loc attribute to return one or more specified row(s).Using it we can replace the value of specified row.
print(df.to_string())
```

For small data sets you might be able to replace the wrong data one by one, but not for big data sets.

To replace wrong data for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries.



```python
for x in df.index:
    if df.loc[x,'Pulse']>130:
        df.loc[x,'Pulse'] = 130
print(df.to_string())
```

Delete rows where "Duration" is higher than 160:

```python
for x in df.index:
  if df.loc[x, "Duration"] > 160:
    df.drop(x, inplace = True)
print(df.to_string())
```

```python
print(df.duplicated()) #To discover duplicates, we can use the duplicated() method.
#The duplicated() method returns a Boolean values for each row
```

```python
df.drop_duplicates(inplace=True)
print(df)
#To remove duplicates, we use the drop_duplicates() method
```

##### Cells with data of wrong format can make it difficult, or even impossible, to analyze data.

##### To fix it, you have two options: remove the rows, or convert all cells in the columns into the same format.
We have column 'Date of birth', and some of them have different format.
If we want to convert all cells in the 'Date of birth' column into dates.Pandas has a to_datetime() method to return them into a same format.


```python
dg['Date of birth'] = pd.to_datetime(dg['Date of birth'], format='mixed')
 
dg
```

##### If that 'Date of birth' column has empty values, we can remove them with dropna() function. To remove empty values of specified column we can use  subset=['Date of birth']


```python
dg.dropna(subset=['Date of birth'], inplace = True)
dg
```
